## Data Cleaning Pipeline using Python

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("raw_business_dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

* convert standerd column name

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

* categorical column missing value fill

```python
df["country"] = df["country"].fillna("Unknown")
df["notes"] = df["notes"].fillna("No Notes")
```

* numeric column missing value handle

```python
df["quantity"] = pd.to_numeric(df["quantity"], errors="coerce")
df["quantity"] = df["quantity"].fillna(df["quantity"].median())
```

```python
df["unit_price"] = pd.to_numeric(df["unit_price"], errors="coerce")
df["unit_price"] = df["unit_price"].fillna(df["unit_price"].median())
```

* drop dupplicate values

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

* Clean Text Columns

```python
df["customer_name"] = df["customer_name"].str.strip().str.title()
df["country"] = df["country"].str.strip().str.title()
df["product_category"] = df["product_category"].str.strip().str.title()
df["sales_rep"] = df["sales_rep"].str.strip().str.title()
```

* Convert Date Column

```python
df["order_date"] = pd.to_datetime(df["order_date"], dayfirst=True)
```

* extra 2 column add

```python
df[["rookie_year", "final_year"]] = df["joined_span"].str.split("-", expand=True)
```

```python
df["rookie_year"] = pd.to_numeric(df["rookie_year"], errors="coerce")
df["final_year"] = pd.to_numeric(df["final_year"], errors="coerce")
```

* drop columns

```python
df = df.drop(columns=["joined_span"])
```

* negative value drop

```python
df = df[df["quantity"] >= 0]
```

* ratting column handling

```python
df["rating"] = df["rating"].str.replace("*", "", regex=False)
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
```

```python
df["rating"] = df["rating"].fillna(df["rating"].median())
```

* career length

```python
df["career_length"] = df["final_year"] - df["rookie_year"]
```

```python
df.head()
```

* gross sales

```python
df['gross_sales'] = df['quantity'] * df['unit_price']
```

* discount ammount

```python
df["discount"] = pd.to_numeric(df["discount"], errors="coerce")
df["discount"] = df["discount"].fillna(df["discount"].median())
```

```python
df["discount_amount"] = df["gross_sales"] * (df["discount"] / 100)
```

* Net Sales

```python
df["net_sales"] = df["gross_sales"] - df["discount_amount"]
```

* Order Year

```python
df["order_year"] = df["order_date"].dt.year
```
